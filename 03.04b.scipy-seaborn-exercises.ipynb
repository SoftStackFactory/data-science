# SciPy Exercise

For this exercise you'll generate some mock data using the SciPy library, then using Seaborn, plot the data using a distplot.

##### Requirements:
- Create a list of sample data using the SciPy stats library.
    - List must contain a minimum of 10,000 random values
    - Values need to be from a uniform distribution
    - Random values must be between 0-10
- Generate Histogram Plot
    - Must use Seaborn visualization library
    

##### Create data

```python
# Import Libraries
from scipy.stats import uniform

def get_data(size=None, loc=None, scale=None):
    """
        Info:
            This function returns a list of n random numbers from a uniform distribution 
        Params:
            size (type: int)
            loc (type: int) 
            scale (type: int)
        Output:
            numpy_array (type: numpy array)
    """
    sample_data = uniform.rvs(size=size, loc=loc, scale=scale)
    return sample_data

size = 10000
loc = 0
scale = 10

# Get sample data
scipy_data = get_data(size=size, loc=loc, scale=scale)
```

##### Histogram Plot using Seaborn

```python
# import seaborn
import seaborn as sns

# Plot Histogram
ax = sns.distplot(scipy_data,
                  bins=100,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})

ax.set(xlabel='Uniform ', ylabel='Frequency')
```

# Seaborn Exercise

In the lesson we covered density plots and briefly mentioned the close relation to Kernel Density Estimation (KDE).

A [KDE Plot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html) is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable.

For this exercise, we're going to use Seaborn to generate a KDE plot for SepalLengthCm using the Iris dataset depicting the probability density for each of the Species.


##### Requirements:
- Import Iris data using Pandas
- Plot the probability density for the Species

##### Import Iris data using Pandas

```python
# Import Pandas
import pandas as pd 

# Import CSV using Pandas
iris_df = pd.read_csv('./data/iris.csv')

print(iris_df['Species'].unique())
iris_df.head()
```

##### KDE Plot using Seaborn

```python
# Import Libraries
import seaborn as sns
from matplotlib import pyplot as plt

# Plotting the KDE Plot for Iris-setosa
sns.kdeplot(iris_df.loc[(iris_df['Species']=='Iris-setosa'), 
            'SepalLengthCm'], color='r', shade=True, Label='Iris-setosa') 

# Plotting the KDE Plot for Iris-vercicolor
sns.kdeplot(iris_df.loc[(iris_df['Species']=='Iris-versicolor'),  
            'SepalLengthCm'], color='b', shade=True, Label='Iris-versicolor') 

# Plotting the KDE Plot for Iris-virginica 
sns.kdeplot(iris_df.loc[(iris_df['Species']=='Iris-virginica'),  
            'SepalLengthCm'], color='g', shade=True, Label='Iris-virginica') 
  
plt.xlabel('Sepal Length') 
plt.ylabel('Probability Density') 
```
